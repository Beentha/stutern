```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("company_sales_data.csv")
data.head()
```

```python
plt.plot(data['month_number'], data['facecream'],   label = 'Face cream Data', marker='o')
plt.plot(data['month_number'], data['facewash'],   label = 'Face wash Data',  marker='o')
plt.plot(data['month_number'], data['toothpaste'], label = 'Toothpaste Data', marker='o')
plt.plot(data['month_number'], data['bathingsoap'], label = 'Bathing soap Data', marker='o')
plt.plot(data['month_number'], data['shampoo'], label = 'Shampoo Data', marker='o')
plt.plot(data['month_number'], data['moisturizer'], label = 'Moisturizer Data', marker='o')

plt.title('Company Sales data')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales Units')
plt.show()
```
