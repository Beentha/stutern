```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('anscombe_quartet_data.csv')
data.head()
```

```python
from turtle import color

plt.figure(figsize=(16, 10))
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
p0, p1 = np.polyfit(data['x1'], data['y1'], 1)
plt.scatter(data['x1'], data['y1'], color='blue')
plt.plot(data['x1'], p0*data['x1']+p1, color = 'blue')
plt.xlabel('x1')
plt.ylabel('y1')  

plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
p2, p3 = np.polyfit(data['x2'], data['y2'], 1)
plt.scatter(data['x2'], data['y2'], color='blue')
plt.plot(data['x2'], p2*data['x2']+p3, color = 'blue') 
plt.xlabel('x2')
plt.ylabel('y2') 

plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
p4, p5 = np.polyfit(data['x3'], data['y3'], 1)
plt.scatter(data['x3'], data['y3'], color='blue')
plt.plot(data['x3'], p4*data['x3']+p5, color = 'blue')
plt.xlabel('x3')
plt.ylabel('y3')  

plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
p6, p7 = np.polyfit(data['x4'], data['y4'], 1)
plt.scatter(data['x4'], data['y4'], color='blue')
plt.plot(data['x4'], p6*data['x4']+p7, color = 'blue')
plt.xlabel('x4')
plt.ylabel('y4')  

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16, 10))

ax1.scatter(data['x1'], data['y1'], color='blue')
ax1.plot(data['x1'], p0*data['x1']+p1, color = 'blue')
ax1.set(xlabel='x1', ylabel='y1')
ax1.set_title('(x1,y1)')
ax2.scatter(data['x2'], data['y2'], color='blue')
ax2.plot(data['x2'], p2*data['x2']+p3, color = 'blue')
ax2.set(xlabel='x2', ylabel='y2')
ax2.set_title('(x2,y2)')
ax3.scatter(data['x3'], data['y3'], color='blue')
ax3.plot(data['x3'], p4*data['x3']+p5, color = 'blue')
ax3.set(xlabel='x3', ylabel='y3')
ax3.set_title('(x3,y3)')
ax4.scatter(data['x4'], data['y4'], color='blue')
ax4.plot(data['x4'], p6*data['x4']+p7, color = 'blue')
ax4.set(xlabel='x4', ylabel='y4')
ax4.set_title('(x4,y4)')

plt.show()
```
