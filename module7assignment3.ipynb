```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
company_sales_data = pd.read_csv('company_sales_data.csv')
company_sales_data.head()
```

```python
# Compare any two products sales data using a bar chart.
fig = plt.figure(figsize=(16, 10))

plt.bar([month-0.25 for month in company_sales_data['month_number']], company_sales_data['facecream'], label = 'Face Cream data', width= 0.45)
plt.bar([month+0.25 for month in company_sales_data['month_number']], company_sales_data['facewash'], label = 'Face Wash data', width= 0.45)

plt.xticks(company_sales_data['month_number'])
plt.title('Facewash and facecream sales data')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.savefig('face_wash_and_face_cream_sales_data.png', dpi=150)

plt.show()
```

```python
# Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale.
fig = plt.figure(figsize=(16, 10))

avg_monthly_sales = [10000 for i in range(len(company_sales_data['month_number']))]
plt.bar(company_sales_data['month_number'], company_sales_data['bathingsoap'])
plt.plot(company_sales_data['month_number'], avg_monthly_sales, linewidth=3.5)

plt.xticks(company_sales_data['month_number'])
plt.title('Bathing Soap Sales Data With The Average Monthly Sales')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.savefig('bathingsoap_sales_data.png', dpi=150)
plt.show()
```

```python
# Calculate the total sales data for the year of each product and represent it on a pie chart
fig = plt.figure(figsize=(16, 10))

sales_data   = [company_sales_data ['facecream'].sum(), company_sales_data ['facewash'].sum(), company_sales_data ['toothpaste'].sum(), 
                company_sales_data ['bathingsoap'].sum(), company_sales_data ['shampoo'].sum(), company_sales_data ['moisturizer'].sum()]
labels = ['Face cream', 'Fase wash', 'Tooth paste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.pie(sales_data, labels=labels, autopct='%0.1f%%')
plt.title('Company Sales data')
plt.legend(loc='upper left')
plt.savefig('sales_data_pie_chart.png', dpi=150)

plt.show()
```
