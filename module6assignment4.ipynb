```python
# An inner merge operation between the orders table and products table
products_orders = pd.merge(orders_df, products_df)
products_orders
```

```python
# Use the resulting table to perform another inner merge operation with the customers table
customer_products_orders = pd.merge(products_orders, customers_df)
customer_products_orders
```

```python
# Perform an inner merge operation between customers and orders tables. 
customer_orders = pd.merge(customers_df, orders_df, on = 'customer_id', how = 'inner')
customer_orders
```

```python
# Perform an outer merge between customers and products tables.
customer_products = pd.merge(customers_df, products_df,  how = 'outer')
customer_products
```

```python
# Perform a left merge between orders and customers
order_customers = pd.merge(orders_df, customers_df,  how = 'left')
order_customers
```

```python
# Perform a right merge between product and customers tables
product_customers = pd.merge(products_df, customers_df,  how = 'right')
product_customers
```
